# Pooled OLS modelling sales with grant

```python
#from pandas.stats.plm import PanelOLS
#reg  = PanelOLS(g,s,time_effects=True)
c.execute('SELECT * FROM jtrain')
j=c.fetchall()
s=np.zeros(len(j))
g=np.zeros(len(j))

for l in range(0,len(j)):
    s[l]=(j[l][5])
    g[l]=(j[l][9])
    
a = np.array(s)
b = np.array(g)
np.vstack((a,b)).T
a[np.invert(np.isnan(a))].shape

sm.OLS(a[np.invert(np.isnan(a))],sm.add_constant(b[np.invert(np.isnan(a))])).fit().summary()
#smf.ols("sm.add_constant(b[np.invert(np.isnan(a))])  ~ a[np.invert(np.isnan(a))]", data = df).fit()
```
